```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
data = pd.read_csv("department.csv")
```

```python
semester = data.iloc[:, 1]
percentage = data.iloc[:, 9]
```

```python
semesters = set(semester)
```

```python
req_percentage = []
count = 0
for i in range(int(len(percentage) / 8)):
    req_per = []
    for j in range(count, count+8):
        req_per.append(percentage[count])
        count += 1
    req_percentage.append(req_per)
```

```python
filtered_data = pd.DataFrame(req_percentage, columns = ("Sem_1", "Sem_2","Sem_3", "Sem_4", "Sem_5", "Sem_6", "Sem_7","Sem_8" ))
```

```python
filtered_data
```

```python
filtered_data.to_csv("Filtered_data.csv")
```

```python
x = filtered_data.iloc[:,0:7]
```

```python
y = filtered_data.iloc[:,7]
```

```python
# dividing the data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
# Getting the Model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# Save the model using pickel
filename = "finalised_model.sav"
pickle.dump(model, open(filename, 'wb'))
```
